```python
##Import the needed library
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## read the dataset 
dataset = pd.read_csv("download-data-for-Aug-2017-Euro-6.csv",encoding='latin1')
dataset.head()
```

```python
#First we Will drop the rows only if all of the values in the row are missing
dataset.dropna(how = 'all',inplace = True)
```

```python
#TO see which have null values
dataset.isna().sum()
```

```python
#drop 
dataset = dataset.drop(['Manufacturer','Model','Description'], axis=1)
```

```python
#drop columns with null values more than 60%
dataset = dataset.drop(['Electric energy consumption Miles/kWh',
                                    'wh/km',
                                    'Maximum range (Km)'
                                    ,'Maximum range (Miles)',
                                    'Electricity cost',
                                    'THC Emissions [mg/km]'
                                    ,'THC + NOx Emissions [mg/km]',
                                    'Particulates [No.] [mg/km]'], axis=1)
```

```python
dataset=dataset.drop_duplicates(keep='first')
```

```python
#dataset.head()
dataset.isna().sum()
```

```python
dataset.iloc[:,[10,11]]
#convert object to string and remove the euro sign
dataset['Fuel Cost 12000 Miles'] = dataset['Fuel Cost 12000 Miles'].str.slice(1,len(dataset['Fuel Cost 12000 Miles'])-1)
#convert the string with ',' to float first remove ',' then convert to float
dataset['Fuel Cost 12000 Miles'] = dataset['Fuel Cost 12000 Miles'].str.replace(",","").astype(float)
```

```python
#convert object to string and remove the euro sign
dataset['Total cost / 12000 miles'] = dataset['Total cost / 12000 miles'].str.slice(1,len(dataset['Total cost / 12000 miles'])-1)
#convert the string with ',' to float first remove ',' then convert to float
dataset['Total cost / 12000 miles'] = dataset['Total cost / 12000 miles'].str.replace(",","").astype(float)
```

```python
#importing sklrn
from sklearn.preprocessing import Imputer
imputer = Imputer(missing_values='NaN', strategy='mean',axis=0)
dataset[['Fuel Cost 12000 Miles', 
                 'Emissions CO [mg/km]',
                 'Engine Capacity',
                 'Metric Urban (Cold)',
                 'Metric Extra-Urban',
                 'Metric Combined',
                'Imperial Urban (Cold)',
                'Imperial Extra-Urban',
                'Imperial Combined',
                'Emissions CO [mg/km]',
                'Emissions NOx [mg/km]']]= imputer.fit_transform(dataset[[
                                         'Fuel Cost 12000 Miles', 
                                         'Emissions CO [mg/km]',
                                         'Engine Capacity',
                                         'Metric Urban (Cold)',
                                         'Metric Extra-Urban',
                                         'Metric Combined',
                                        'Imperial Urban (Cold)',
                                        'Imperial Extra-Urban',
                                        'Imperial Combined',
                                        'Emissions CO [mg/km]',
                                        'Emissions NOx [mg/km]']])
```

```python
dataset.isna().sum()
```

```python
dataset['Transmission'].fillna(method='ffill', inplace=True)
```

```python
dataset.isna().sum()
```

```python
# import the needed librarys
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
dataset['Transmission'] = encoder.fit_transform(dataset['Transmission'])
```

```python
encoder = LabelEncoder()
dataset['Fuel Type'] = encoder.fit_transform(dataset['Fuel Type'])
```

```python
corr = dataset.corr()
fig, ax = plt.subplots(figsize=(10, 10))
ax.matshow(corr)
plt.xticks(range(len(corr.columns)), corr.columns, rotation='vertical')
plt.yticks(range(len(corr.columns)), corr.columns)
plt.grid(True)
plt.show()
```

```python
## separate the features and the target columns
features_matrix = dataset.drop('Emissions CO [mg/km]', axis=1)
features_matrix
```

```python
goal_vector = dataset['Emissions CO [mg/km]']
goal_vector
```

```python
# import the modules and divide the dataset into training and testing set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features_matrix, goal_vector, test_size = 0.2, random_state = 0)
```

```python
#fitting mulitiple linear re3grission model to TRAINNING set:
from sklearn.linear_model import LinearRegression
#create opbject of linear regression model
model_fit = LinearRegression()
model_fit.fit(x_train,y_train)
```

```python
# predicting the test set result
y_pred = model_fit.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)#r2_score is getting the diffrence between y_test(the acual result) and y_predict
```

```python
model_fit.score(x_test, y_test)
for col in x_test:
    y_pred = model_fit.predict(x_test)
    plt.plot(x_test.loc[:, col], y_test, '.', label='actual')
    plt.plot(x_test.loc[:, col], y_pred, '.', label='pred')
    plt.legend()
    plt.title(f"Emissions CO againist {col}")
    plt.show()
```
